# Simple Linear Regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
plt.figure(figsize=(10,7))
x=df[['NA_Sales']] #independant variable
y=df[['Global_Sales']] #dependant variable
model=LinearRegression()
model.fit(x,y) #fitting the data in linear regression model
ypred=model.predict(x)
plt.scatter(x,y)
plt.plot(x,ypred,color='r',marker='o')
model.score(x,y)
print("Y intercept=",model.intercept_)
print("slope=",model.coef_)
```

## No autocorrelation of residuals

Checking for autocorrelation To ensure the absence of autocorrelation we use Ljungbox test.
Null Hypothesis: Autocorrelation is absent.
Alternative Hypothesis: Autocorrelation is present.

```python
from statsmodels.stats import diagnostic as diag
min(diag.acorr_ljungbox(res , lags = 40)[1])
```

### Since p value is less than 0.05 we reject the null hypothesis that error terms are not autocorrelated.


```python
# autocorrelation
import statsmodels.api as sm

sm.graphics.tsa.plot_acf(res, lags=40)
plt.show()
```

## partial autocorrelation

```python
# partial autocorrelation
sm.graphics.tsa.plot_pacf(res, lags=40)
plt.show()
```
